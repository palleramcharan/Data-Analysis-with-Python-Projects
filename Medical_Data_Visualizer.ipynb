```python
import pandas as pd
import io
import matplotlib.pyplot as plt

df=pd.read_csv(io.BytesIO(uploaded['medical_examination.txt']))
df
```

```python
df['overweight']=((df['weight'])/((df['height']/100)**2)).apply(lambda x: 1 if x > 25 else 0)

df['cholesterol']=df['cholesterol'].apply(lambda x:0 if x==1 else 1)
df['gluc']=df['gluc'].apply(lambda x:0 if x==1 else 1)
df
```

```python
df_cat=pd.melt(df,id_vars=['cardio'],value_vars=['cholesterol', 'gluc', 'smoke', 'alco', 'active','overweight'])
df_cat['total']=1
df_cat=df_cat.groupby(['cardio','variable','value'],as_index=False).count()
df_cat
```

```python
import seaborn as sns
fig=sns.catplot(data=df_cat,x='variable',y='total',hue='value',kind="bar",col='cardio')
```

```python
df_heat=df[(df['ap_lo'] <= df['ap_hi']) &
           (df['height'] >= df['height'].quantile(0.025)) &
           (df['height'] <= df['height'].quantile(0.975)) &
           (df['weight'] >= df['weight'].quantile(0.025)) &
           (df['weight'] <= df['weight'].quantile(0.975))]
df_heat
```

```python
corr=df_heat.corr(method='pearson')
corr
```

```python
import numpy as np
mask=np.triu(corr) #getting the upper triangle
mask
```

```python
f, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(corr,linewidths=1,annot=True, fmt=".1f",square=True,mask=mask,center=0.08,cbar_kws={'shrink':0.5})
```
